# **Introdução ao XGBoost (*Extreme Gradient Boosting*)**

Historicamente, algoritmos de Machine Learning dos anos 70 e 80 (Regressão Linear, Naive Bayes) lidavam com tipos de dados específicos. Nos anos 90, surgiram opções mais potentes, como Random Forest e o próprio Gradient Boosting, mas que ainda sofriam com *overfitting* e grande perda de velocidade ao lidar com bases de dados gigantes.

O `XGBoost` é veio para mitigar esse problema. Ele é uma biblioteca de aprendizado de máquina focada em algoritmos de gradient boosting baseados em árvores. 

O `XGBoost` é uma implementação avançada e otimizada do *gradient boosting*, que cria modelos fortes combinando vários "aprendizes fracos", onde cada nova árvore corrige os erros da anterior. Utiliza técnicas para processamento paralelo, tornando o treinamento muito rápido. Incorpora regularização (parâmetros L1 e L2) para evitar *overfitting*, resultando em modelos mais generalizáveis.

### **Exemplo Prático Utilizando XGBoost**

#### **Classificação**

```python
# XGBoost:
from xgboost import XGBClassifier

# Dataset:
from sklearn.datasets import load_breast_cancer

# Treino e Teste:
from sklearn.model_selection import train_test_split, cross_val_score

# Normalizador:
from sklearn.preprocessing import StandardScaler

# Metricas de avaliação:
from sklearn.metrics import accuracy_score
```

```python
# Carregando a base de dados:
data = load_breast_cancer()

X = data.data
y = data.target

print(f'X = {X.shape}')
print(f'y = {y.shape}')
```

```python
# Visualizando algumas carcterísticas do dataset:

print(f'Classes: {data.target_names}\n')

print(f'Features: {data.feature_names}\n')

print(f'Quantidade de valores em cada classe: malignant = {list(y).count(0)} | beningn = {list(y).count(1)}')
```

```python
# Dividindo em teste e treino:

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=42)

print(f'X_train shape = {X_train.shape}')
print(f'y_train shape = {y_train.shape}')
print(f'X_test shape = {X_test.shape}')
print(f'y_test shape = {y_test.shape}')
```

```python
# Normalizando os valores:

std_scaler = StandardScaler()

std_scaler.fit(X_train)

X_train = std_scaler.transform(X_train)
X_test = std_scaler.transform(X_test)
```

```python
# Instânciando o modelo e ajustando:

xgb_classifier = XGBClassifier(n_estimators = 5, max_depth = 5, learning_rate = 0.5, objective='binary:logistic')

scores = cross_val_score(xgb_classifier, X_train, y_train)

for idx, score in enumerate(scores):

    print(f'Score {idx+1}: {score:.2%}')

print(f'Mean = {scores.mean():.2%}')
```

```python
xgb_classifier.fit(X_train, y_train)
```

```python
# Avaliando o modelo

predict_test = xgb_classifier.predict(X_test)
acc = accuracy_score(y_test, predict_test)

print(f'acc = {acc:.2%}')
```

#### **Regressão**

```python
# XGBoost:
from xgboost import XGBRegressor

# Dataset:
from sklearn.datasets import fetch_california_housing

# Treino e Teste:
from sklearn.model_selection import train_test_split

# Normalizador:
from sklearn.preprocessing import StandardScaler

# Metricas de avaliação:
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Carregando o dataset:
data = fetch_california_housing()

X = data.data
y = data.target

print(f'X = {X.shape}')
print(f'y = {y.shape}')
```

```python
# verificando a correlação linear dos dados:

import numpy as np
import pandas as pd

data_df = pd.DataFrame(X)
data_df.columns = [data.feature_names]

data_df[data.target_names[0]] = y

data_df
```

```python
data_df.corr()
```

```python
# Dividindo em teste e treino:

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=42)

print(f'X_train shape = {X_train.shape}')
print(f'y_train shape = {y_train.shape}')
print(f'X_test shape = {X_test.shape}')
print(f'y_test shape = {y_test.shape}')
```

```python
# Normalizando os dados:
std_scaler = StandardScaler()
std_scaler.fit(X_train, y_train)

X_train = std_scaler.transform(X_train)
X_test = std_scaler.transform(X_test)
```

```python
# Instânciando o modelo e ajustando:

# Existe uma relação entre n_estimators e learning_rate

xgb_regressor = XGBRegressor(n_estimators = 200, max_depth = 5, learning_rate = 0.35, verbosity = 3)

xgb_regressor.fit(X_train, y_train)
```

```python
# Analisando 

predict_ = xgb_regressor.predict(X_test)

mse = mean_squared_error(y_test, predict_)

print(f'Root Mean Squared Error: {np.sqrt(mse)}')
```

```python
r2_score(y_test, predict_)
```
